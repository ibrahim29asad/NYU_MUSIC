```python
import os

# --- 1. Create model_rnn.py ---
model_code = """
import torch
import torch.nn as nn
from torch.nn import functional as F

class RNNConfig:
    def __init__(self, vocab_size, n_layer=1, n_embd=128, block_size=256, device='cpu', **kwargs):
        self.vocab_size = vocab_size
        self.n_layer = n_layer
        self.n_embd = n_embd
        self.block_size = block_size
        self.device = device
        self.dropout = 0.0
        self.bias = False

class RNN(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.config = config
        self.token_embedding = nn.Embedding(config.vocab_size, config.n_embd)
        self.lstm = nn.LSTM(
            input_size=config.n_embd,
            hidden_size=config.n_embd,
            num_layers=config.n_layer,
            batch_first=True,
            dropout=0.0
        )
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size)
        print(f"RNN Model Created: {self.get_num_params()/1e6:.2f}M params")

    def get_num_params(self, non_embedding=True):
        n_params = sum(p.numel() for p in self.parameters())
        if non_embedding:
            n_params -= self.token_embedding.weight.numel()
        return n_params

    def forward(self, idx, targets=None):
        x = self.token_embedding(idx)
        x, _ = self.lstm(x)
        logits = self.lm_head(x)
        loss = None
        if targets is not None:
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        return logits, loss

    def crop_block_size(self, block_size):
        pass
"""

# Write it to the current directory
with open("model_rnn.py", "w") as f:
    f.write(model_code)
print(" Created model_rnn.py")

# --- 2. Create train_rnn.py (The corrected version) ---
train_code = """
import os
import time
import pickle
from contextlib import nullcontext
import numpy as np
import torch
from model_rnn import RNNConfig, RNN

# --- Configuration ---
out_dir = 'out'
eval_interval = 250
log_interval = 10
eval_iters = 20
eval_only = False
always_save_checkpoint = True # Force saving
init_from = 'scratch'
dataset = 'music_abc'
gradient_accumulation_steps = 1

batch_size = 64
block_size = 256
n_layer = 4
n_embd = 128
learning_rate = 1e-3
max_iters = 5000
device = 'cuda'
compile = False

# Read args
config_keys = [k for k,v in globals().items() if not k.startswith('_') and isinstance(v, (int, float, bool, str))]
try:
    exec(open('configurator.py').read())
except FileNotFoundError:
    try:
        exec(open('nanoGPT/configurator.py').read())
    except:
        pass

config = {k: globals()[k] for k in config_keys}

# --- CRITICAL FIX: Create the output directory ---
if not os.path.exists(out_dir):
    os.makedirs(out_dir, exist_ok=True)
    print(f"Created output directory: {out_dir}")

torch.manual_seed(1337)
ctx = nullcontext() if 'cpu' in device or 'mps' in device else torch.amp.autocast(device_type='cuda', dtype=torch.float16)

# Data Path Logic
if os.path.exists(os.path.join('data', dataset)):
    data_dir = os.path.join('data', dataset)
elif os.path.exists(os.path.join('nanoGPT', 'data', dataset)):
    data_dir = os.path.join('nanoGPT', 'data', dataset)
else:
    data_dir = os.path.join('data', dataset)

train_data = np.memmap(os.path.join(data_dir, 'train.bin'), dtype=np.uint16, mode='r')
val_data = np.memmap(os.path.join(data_dir, 'val.bin'), dtype=np.uint16, mode='r')

def get_batch(split):
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([torch.from_numpy((data[i:i+block_size]).astype(np.int64)) for i in ix])
    y = torch.stack([torch.from_numpy((data[i+1:i+1+block_size]).astype(np.int64)) for i in ix])
    if 'cuda' in device:
        x, y = x.pin_memory().to(device, non_blocking=True), y.pin_memory().to(device, non_blocking=True)
    else:
        x, y = x.to(device), y.to(device)
    return x, y

# Model Setup
iter_num = 0
best_val_loss = 1e9
model_args = dict(n_layer=n_layer, n_embd=n_embd, block_size=block_size, vocab_size=50304)
if os.path.exists(os.path.join(data_dir, 'meta.pkl')):
    with open(os.path.join(data_dir, 'meta.pkl'), 'rb') as f:
        meta = pickle.load(f)
    model_args['vocab_size'] = meta['vocab_size']

gptconf = RNNConfig(**model_args)
model = RNN(gptconf)
model.to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)

# Training Loop
X, Y = get_batch('train')
t0 = time.time()

while True:
    # Evaluation & Saving
    if iter_num % eval_interval == 0:
        model.eval()
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):
            X_batch, Y_batch = get_batch('val')
            with ctx:
                _, loss = model(X_batch, Y_batch)
            losses[k] = loss.item()
        val_loss = losses.mean()
        print(f"step {iter_num}: val loss {val_loss:.4f}")

        # --- CRITICAL FIX: Save the Checkpoint ---
        if val_loss < best_val_loss or always_save_checkpoint:
            best_val_loss = val_loss
            if iter_num > 0:
                checkpoint = {
                    'model': model.state_dict(),
                    'optimizer': optimizer.state_dict(),
                    'model_args': model_args,
                    'iter_num': iter_num,
                    'best_val_loss': best_val_loss,
                    'config': config,
                }
                print(f"saving checkpoint to {out_dir}")
                torch.save(checkpoint, os.path.join(out_dir, 'ckpt.pt'))
        model.train()

    # Forward Pass
    with ctx:
        _, loss = model(X, Y)

    X, Y = get_batch('train')
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    optimizer.step()
    optimizer.zero_grad(set_to_none=True)

    # Logging
    if iter_num % log_interval == 0:
        dt = time.time() - t0
        t0 = time.time()
        print(f"iter {iter_num}: loss {loss.item():.4f}, time {dt*1000:.2f}ms")

    iter_num += 1
    if iter_num > max_iters:
        break
"""

with open("train_rnn.py", "w") as f:
    f.write(train_code)
print(" Created train_rnn.py")
```
